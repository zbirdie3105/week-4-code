```python
df['is_female'] = pd.get_dummies(df.Sex)['female']
X = df.drop(columns=['PassengerId', 'Survived', 'Name', 'Sex'])
X.head()
```

```python
X['ager'] = pd.cut(X.Age, 3, labels=[1,2,3])
X = X.drop(columns=['Age'])
X.head()
```

```python
X['fam_size'] = X.SibSp + X.Parch
X = X.drop(columns=['SibSp', 'Parch'])
X.head()
```

```python
X.Embarked.value_counts()
```

```python
X['Embarked'] = pd.Categorical(X.Embarked, ordered=True).codes
X.head()
```

```python
X = X.drop(columns=['Ticket'])
X.head()
```

```python
X['Cabin'] = X.Cabin.map(lambda c: 0 if pd.isna(c) else ord(c[0]))
X.head()
```
